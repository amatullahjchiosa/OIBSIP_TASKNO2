#**Upload Data**

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
%matplotlib inline
```

```python
df=pd.read_csv('/content/Unemployment_Rate_upto_11_2020.csv')
df.head(267)
```

```python
#Assess how many columns and row this dataset has.
df.shape
```

```python
#Assess which entries are floats and objects in this dataset.
df.info()
```

```python
#Obtain a summarised description of the dataset.
df['Date']=pd.to_datetime(df[' Date'])
df.describe()
```

##**Unemployment Rate over time in Indian Populations**

```python
#Using a figure, visualise the rise and fall of unemployment rates in the overall Indian populations.
plt.figure(figsize=(15,5))
sns.lineplot(data=df, x=' Date',y=' Estimated Unemployment Rate (%)')
```

##**Unemployment Rate during Coronavirus in India**


```python
covid_19=df[df.Date>='01-01-2020']
covid_19.head()
```

```python
#Assess the effect of coronavirus on the unemployment rate in the overall Indian Population.
plt.figure(figsize=(15,5))
sns.lineplot(data=covid_19, x=' Date', y=' Estimated Unemployment Rate (%)')
```

#**Visual assessment of Unemployment rate in the different regions of India**

```python
#Plotting a heatmap to check the correlation between feutures of the dataset
dfplot=sns.heatmap(df.corr())
```

```python
#plotting a histogram to find the correlation between unemployment rate and different regions in India
plt.figure(figsize=(15,5))
sns.histplot(x=' Estimated Unemployment Rate (%)', hue='Region.1', data=df)
plt.show()
```
